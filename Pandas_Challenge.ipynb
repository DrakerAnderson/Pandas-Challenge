```python
import pandas as pd
from pathlib import Path
```

```python
# Load CSV Data for Schools and Students
school_data_path = Path('Resources', 'schools_complete.csv')
school_data = pd.read_csv(school_data_path)

student_data_path = Path('Resources', 'students_complete.csv')
student_data = pd.read_csv(student_data_path)

# Merge data sets
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

# District Summary

```python
# Total number of unique schools

total_schools = len(school_data_complete['school_name'].unique())

total_schools
```

```python
# Total students
total_students = school_data_complete['Student ID'].count()

total_students
```

```python
# Total budget
total_budget = school_data['budget'].sum()

total_budget
```

```python
# Average math score
avg_math_score = school_data_complete['math_score'].mean()

avg_math_score
```

```python
# Average reading score
avg_reading_score = school_data_complete['reading_score'].mean()

avg_reading_score
```

```python
# Percentage of students who passed math
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(total_students) * 100
print("Percent of Students passing math:", round(passing_math_percentage,2))
```

```python
# Percentage of students who passed reading
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(total_students) * 100
print("Percent of Students passing reading:", round(passing_reading_percentage,2))
```

```python
# Percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(total_students) * 100
print("Percent of Students Passing overall:", round(overall_passing_rate,2))
```

```python
district_summary = pd.DataFrame(
	{
		"Total Schools": [total_schools],
		"Total Students": [total_students],
		"Total Budget": [total_budget],
		"Average Math Score": [avg_math_score],
		"Average Reading Score": [avg_reading_score],
		"% Passing Math": [passing_math_percentage],
		"% Passing Reading": [passing_reading_percentage],
		"% Overall Passing": [overall_passing_rate],
	}
)

#Format the data
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Print the summary 
district_summary
```

```python
# Perform the necessary calculations to create a DataFrame that lists the average math score for students of each grade level (9th, 10th, 11th, 12th) at each school.

# Group the data by school name and grade, and calculate the average math score
avg_math_scores_by_grade = student_data.groupby(['school_name', 'grade'])['math_score'].mean().reset_index()

# Pivot the data to create a separate column for each grade level
avg_math_scores_by_grade = avg_math_scores_by_grade.pivot(index='school_name', columns='grade', values='math_score')

# Rename the columns to make them more readable
avg_math_scores_by_grade.columns = [f'Average Math Score - {grade}' for grade in avg_math_scores_by_grade.columns]

# Reset the index to create a clean DataFrame
avg_math_scores_by_grade = avg_math_scores_by_grade.reset_index()

avg_math_scores_by_grade.head()
```

```python
# Perform the necessary calculations to create a DataFrame that lists the average reading score for students of each grade level (9th, 10th, 11th, 12th) at each school.

# Group the data by school name and grade, and calculate the average reading score
avg_reading_scores_by_grade = student_data.groupby(['school_name', 'grade'])['reading_score'].mean().reset_index()

# Pivot the data to create a separate column for each grade level
avg_reading_scores_by_grade = student_data.pivot_table(index='school_name', columns='grade', values='reading_score', aggfunc='mean')

# Rename the columns to make them more readable
avg_reading_scores_by_grade.columns = [f'Average reading score - {grade}' for grade in avg_reading_scores_by_grade.columns]

# Reset the index to create a clean DataFrame
avg_reading_scores_by_grade = avg_reading_scores_by_grade.reset_index()

avg_reading_scores_by_grade.head()
```

```python
# Create bins for school spending
spending_bins = [0, 585, 630, 645, 680]

# Create labels for the bins
labels = ["<$584", "$585-629", "$630-644", "$645-675"]
```

```python
#Copy the per school summary data to a new dataframe
school_spending_df = per_school_summary.copy()
```

```python

#Convert the per student budget to a float
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].replace('[\$,]', '', regex=True).astype(float)
# school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].str.replace('[\$,]', '', regex=True).astype(float)
```

```python
#Pd.cut to group the data
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], spending_bins, labels=labels)

#Show the data
school_spending_df
```

```python
# Calculate the averages for the spending bins
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Percent Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Percent Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Percent Overall Passing"].mean()
```

```python
#Create a new dataframe to hold the above results
spending_summary = pd.DataFrame(
	{
		"Average Math Score": spending_math_scores,
		"Average Reading Score": spending_reading_scores,
		"Percent Passing Math": spending_passing_math,
		"Percent Passing Reading": spending_passing_reading,
		"Percent Overall Passing": overall_passing_spending,
	}
)

#Show the data
spending_summary
```

```python
#Create bins for school size
size_bins = [0, 1000, 2000, 5000]

#Create labels for the bins
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
#Pd.cut to group the data
per_school_summary ["School Size"] = pd.cut(per_school_summary["Total Students"], size_bins, labels=labels)

#Show the data
per_school_summary
```

```python
# Calculate averages for the desired columns.
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["Percent Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["Percent Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["Percent Overall Passing"].mean()	 
```

```python
#Create a new dataframe to hold the above results
size_summary = pd.DataFrame(
	{
		"Average Math Score": size_math_scores,
		"Average Reading Score": size_reading_scores,
		"% Passing Math": size_passing_math,
		"% Passing Reading": size_passing_reading,
		"% Overall Passing": size_overall_passing,
	}
)

#Show the data
size_summary
```
